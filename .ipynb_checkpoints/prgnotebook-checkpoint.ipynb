```python
# Import all libraries being used
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

# df or dataframe is the variable that will store the data from the penguins_size.csv file.
df = pd.read_csv('penguins_size.csv')
df.rename(columns={'species' : 'Species', 'island' : 'Island', 'culmen_length_mm' : 'CulmenLength', 'culmen_depth_mm' : 'CulmenDepth', 'flipper_length_mm' : 'FlipperLength', 'body_mass_g' : 'BodyMass', 'sex' : 'Sex'}, inplace=True, errors='raise')
df.drop(columns=['Island'], inplace=True)
df = df.dropna()
df = df[df['Sex'].isin(['MALE', 'FEMALE'])]

print(df)
```

```python
# The dataframe is divided based on the specific species.
adelie_df = df[df['Species'] == 'Adelie']
chinstrap_df = df[df['Species'] == 'Chinstrap']
gentoo_df = df[df['Species'] == 'Gentoo']

# Create pairplot variable for the entire dataframe with emphasis on the differing species.
df_plot = sb.pairplot(df, hue='Species')

# Create pairplot variables with emphasis on the penguin's sex are created for the species-specific dataframes.
adelie_plot = sb.pairplot(adelie_df, hue='Sex')
chinstrap_plot = sb.pairplot(chinstrap_df, hue='Sex')
gentoo_plot = sb.pairplot(gentoo_df, hue='Sex')

# Set the titles for each plot.
df_plot.fig.suptitle('Penguin Pairplot', y=1.01, fontsize=20, fontweight='bold')
adelie_plot.fig.suptitle('Adelie Species Pairplot', y=1.01, fontsize=20, fontweight='bold')
chinstrap_plot.fig.suptitle('Chinstrap Species Pairplot', y=1.01, fontsize=20, fontweight='bold')
gentoo_plot.fig.suptitle('Gentoo Species Pairplot', y=1.01, fontsize=20, fontweight='bold')
```

```python
# Create a violin plot variable with culmen length and species.
sb.violinplot(x='Species', y='CulmenLength', data=df)
plt.title('Culmen Length By Species', y=1.01, fontsize=15, fontweight='bold')
```

```python
# Create dataset variables to isolate the target.
dfx = df[['CulmenLength', 'CulmenDepth', 'FlipperLength', 'BodyMass', 'Sex']]
dfy = df['Species']

# Transform dfx by encoding the 'Sex' column in dfx since it is categorical.
ctransform = ColumnTransformer([('encoder', OneHotEncoder(), ['Sex'])], remainder='passthrough')
dfx_transformed = ctransform.fit_transform(dfx)

# Split the data in the dataframe for training and testing.
trainDFX, testDFX, trainDFY, testDFY = train_test_split(dfx_transformed, dfy, test_size=0.2, random_state=40)

# Train the SVM classification model with the training data.
svm_model = SVC(kernel='linear')
svm_model.fit(trainDFX, trainDFY)

# Test the SVM classification model with the testing data.
predictDFY = svm_model.predict(testDFX)
print(classification_report(testDFY, predictDFY))
```

```python
confusion_matrix = confusion_matrix(testDFY, predictDFY)
confusionDF = pd.DataFrame(confusion_matrix, index=['Actual Adelie', 'Actual Chinstrap', 'Actual Gentoo'], columns=['Predicted Adelie', 'Predicted Chinstrap', 'Predicted Gentoo'])

plt.figure(figsize = (8, 6))
sb.heatmap(confusionDF, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix', fontsize=15, fontweight='bold')
plt.xlabel('Predicted Species')
plt.ylabel('Actual Species')
plt.show()
```
